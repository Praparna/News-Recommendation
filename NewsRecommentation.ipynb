```python
import requests
from bs4 import BeautifulSoup
url = "https://www.opindia.com/latest-news/page/5/"
data = requests.get(url)
soup = BeautifulSoup(data.content,"html.parser")
```

```python
headlines = []
for i in soup.find_all("h3",class_="entry-title td-module-title"):
   headlines.append(i.find("a")["title"])
```

```python
urllist = []
for i in range(1,11,1):
   url = "https://www.opindia.com/latest-news/page/" + str(i)
   urllist.append(url)
headlines = []
for j in urllist:
   data = requests.get(j)
   soup = BeautifulSoup(data.content,"html.parser")
   for i in soup.find_all("h3",class_="entry-title td-module-title"):
       headlines.append(i.find("a")["title"])
```

```python
import pandas as pd
```

```python
A = pd.DataFrame()
A['headlines'] = headlines
```

```python
A
```

```python
from preprocessing_NLP import rem_spl_chars,sw,tokenize,preproc,stem,lem
```

```python
q = []
for i in A.headlines:
   for j in (sw(lem(stem(tokenize(rem_spl_chars(i)))))):
       q.append(j)
```

```python
q
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
tf = TfidfVectorizer()
X = tf.fit_transform(q)
```

```python
from sklearn.cluster import KMeans
km = KMeans(n_clusters=9)
model = km.fit(X)
B = pd.DataFrame(model.labels_,columns=["cluster"])
```
